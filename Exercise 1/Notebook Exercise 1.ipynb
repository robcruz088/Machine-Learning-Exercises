### 2.1

```python
diabetesShape = diabetes_df.shape
diabetesAxes = diabetes_df.axes
diabetesDtype = diabetes_df.dtypes

print("The shape of the dataset: ", diabetesShape)
print("\nThe axes of the dataset: \n", diabetesAxes)
print("\nThe dtype of the dataset: \n", diabetesDtype)
```

This is *supervised learning* problem. The first 8 features are independent, with the Outcome being a dependent feature on those previous ones. It is trying to predict someone having diabetes based on those various features. 

This dataset contains 8 independent variables.
<ol>
    <li>Pregnancies</li>

Only 1 categorical data variable and it is *Outcome*, which uses **label encoding**

```python
young = 0
middle = 0
old = 0
for i in diabetes_df['Age']:
    if(i >= 21 and i <= 40):
        young +=1 
    elif (i >= 41 and i <= 60):
        middle += 1
    elif (i >= 61):
        old += 1
    else:
        continue
        
print("Young adults (21-40): ",young)
print("Middle-aged adults (41-60): ",middle)
print("Old-aged adults (61+): ",old)

print("\nAll ages add up to: ",young + middle + old)
```

### 2.2

```python
diabetes_df['Glucose'].replace(0, np.nan, inplace = True)
diabetes_df['BloodPressure'].replace(0, np.nan, inplace = True)
diabetes_df['SkinThickness'].replace(0, np.nan, inplace = True)
diabetes_df['Insulin'].replace(0, np.nan, inplace = True)
diabetes_df['BMI'].replace(0, np.nan, inplace = True)

# checking for any negative values
for i in diabetes_df.iloc[1,0:7]: # select "first" row (the row without column name) and columns 0-7
    if(i < 0):
        diabetes_df.replace(i, np.nan, inplace = True)
        
diabetes_df.head(20)
```

For Glucose, Blood Pressure, Skin Thickness, Insuling and BMI, I concluded that any valid value in those features could not be 0. Therefore, they were replaced with *NaN*. Pregnancies can have valid values of 0 so those were left alone. Due to being unfamiliar with Diabetes Pedigree Function, it was also not checked for 0 values since those could possibly be valid. </br>

Age did not need to be touched since there are no values of Age less than 21, this was checked in the previous cells when accounting for Age ranges and they all add up to 768. </br>

Furthermore, since I cannot see every value, I also decided to check for any non-positive value that could be sneaking around in the dataset. Non-positive values do not have any place in this dataset 

### 2.3

```python
# Normalizing the dataset
diabetes_mean = diabetes_df.loc[:,'Pregnancies':'Age'] - diabetes_df.loc[:,'Pregnancies':'Age'].mean(skipna = True)
diabetes_var = diabetes_mean / (diabetes_mean.std(skipna = True))

diabetes_df['Pregnancies'] = diabetes_var['Pregnancies'] 
diabetes_df['Glucose'] = diabetes_var['Glucose'] 
diabetes_df['BloodPressure'] = diabetes_var['BloodPressure'] 
diabetes_df['SkinThickness'] = diabetes_var['SkinThickness'] 
diabetes_df['Insulin'] = diabetes_var['Insulin'] 
diabetes_df['BMI'] = diabetes_var['BMI'] 
diabetes_df['DiabetesPedigreeFunction'] = diabetes_var['DiabetesPedigreeFunction'] 
diabetes_df['Age'] = diabetes_var['Age'] 

diabetes_df.head(20)
```

### 2.4

```python
# One Hot Encoded dataframe
diabetes_df2 = diabetes_df
diabetes_df2['Outcome'] = diabetes_df2['Outcome'].astype('object')
killer_queen = 0
for i in diabetes_df2['Outcome']:
    if (i == 0 ):
        diabetes_df2.at[killer_queen,'Outcome'] = [1,0]
        killer_queen += 1
    elif (i == 1):
        diabetes_df2.at[killer_queen,'Outcome'] = [0,1]
        killer_queen += 1

diabetes_df.head(20)

```

### 2.5

```python
# For outcome with values 0
zero_outcome = diabetes_df.loc[diabetes_df['Outcome'] == 0]
zero_median = zero_outcome.median(skipna = True)
zero_replace = zero_outcome.fillna(zero_median)

# For outcome with value 1
one_outcome = diabetes_df.loc[diabetes_df['Outcome'] == 1]
one_median = one_outcome.median(skipna = True)
one_replace = one_outcome.fillna(one_median)

# Combining the two separate datasets
diabetes_combine = [zero_replace, one_replace]
diabetes_median = pd.concat(diabetes_combine)
final_diabetes_dataset = diabetes_median.sort_index(axis = 0)
final_diabetes_dataset
```

```python
# One hot encoding the last dataset
final_diabetes_dataset['Outcome'] = final_diabetes_dataset['Outcome'].astype('object')
killer_queen = 0
for i in final_diabetes_dataset['Outcome']:
    if (i == 0 ):
        final_diabetes_dataset.at[killer_queen,'Outcome'] = [1,0]
        killer_queen += 1
    elif (i == 1):
        final_diabetes_dataset.at[killer_queen,'Outcome'] = [0,1]
        killer_queen += 1

final_diabetes_dataset.to_csv('diabetes_dataset_processed.csv', encoding='UTF-8', index=False) # Export to csv
```
